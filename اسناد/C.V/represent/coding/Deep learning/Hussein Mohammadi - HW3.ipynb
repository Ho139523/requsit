## we make a simple NN with 5 hidden layers and one output layer

### Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### prepare the data

```python
x = np.arange(20).reshape(20, 1)
y = np.sin(x).reshape(20, 1)
print(f'The shape of the inputs are {x.shape}')
print(f'The shape of the outputs are {y.shape}')
```

```python
x
```

```python
y
```

### define the number of nodes in each layer

```python
n_x = 1 # size of the input layer
n_h = 5 # size of the hidden layer
n_y = 1 # size of the output layer 
```

### initialize the weights and bias randomly

```python
w_1 = np.random.randn(n_x, n_h) * 0.01
b_1 = np.random.randn(1, n_h) * 0.01

w_2 = np.random.randn(n_h, n_y) * 0.01
b_2 = np.random.randn(1,n_y) * 0.01
```

```python
w_1, b_1, w_2, b_2
```

### define the sigmoid and tanh activation functions

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def tanh(z):
    return (1 - np.exp(-2 * z))/(1 + np.exp(-2 * z))
```

### define the derivative of the sigmoid and tanh functions

```python
def sigmoid_derevative(z):
    return np.exp(-z) / ((1+np.exp(-z))**2)

def tang_derivative(z):
    return 2 * np.exp(-2 * z)/(1 + np.exp(-4 * z) + 2 * np.exp(-2 * z))
```

### define the forward propagation

```python
def forward_prop(x, w_1, w_2):
    z1 = np.dot(x, w_1) + b_1
    a1 = sigmoid(z1)
    
    z2 = np.dot(a1, w_2) + b_2
    y_hat = tanh(z2)
    
    return z1, a1, z2, y_hat
```

```python
x.reshape(1,20)
```

### define the backward propagation

```python
def backward_prop(y_hat, z1, a1, z2):
    delta2 = np.multiply(-(y-y_hat), tang_derivative(z2))
    dw2 = np.dot(a1.T, delta2)
    
    delta1 = np.dot(delta2,w_2.T) * sigmoid_derevative(z1)
    dw1 = np.dot(x.T, delta1)
    
    return dw1, dw2, delta2, delta1
```

### define the cost function

```python
def cost_function(y, y_hat):
    j = 0.5 * sum((y-y_hat)**2) # MSE
    return j
```

### set the learning rate and the number of the training iterations

```python
alpha = 0.01 # the learning rate
n_iteration = 10000 # number of the iterations
```

### Now, Start the training the network

```python
cost = []
for i in range(n_iteration):
    z1, a1, z2, y_hat = forward_prop(x, w_1, w_2)
    dw1, dw2, delta2, delta1= backward_prop(y_hat, z1, a1, z2)
    
    w_1 = w_1 - alpha * dw1
    b_1 = b_1 - alpha * delta1
    w_2 = w_2 - alpha * dw2
    b_2 = b_2 - alpha * delta2
    
    c = cost_function(y, y_hat)
       
    cost.append(c)

```

### Now, plot the cost function

```python
plt.grid()

plt.plot(range(n_iteration), cost)

plt.title("Cost function")
plt.xlabel('Training Iteration')
plt.ylabel('Cost')
```
